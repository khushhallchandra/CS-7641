```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
```

```python
z = pd.read_csv('./mnist_pca_nn.csv')
plt.figure(1)
plt.title("Neural Network Inference Time")
plt.xlabel("Number of Features")
plt.ylabel("Time(s)")

plt.grid()

x = [36, 76, 116, 156, 196]
plt.plot(x, z['mean_score_time'][0:5], 'o--',color='r',label="Layer=1")
plt.plot(x, z['mean_score_time'][5:10], 'o--',color='g',label="Layer=2")
plt.plot(x, z['mean_score_time'][10:15], 'o--',color='b',label="Layer=3")
plt.plot(x, z['mean_score_time'][15:20], 'o--',color='y',label="Layer=4")
plt.plot(x, z['mean_score_time'][20:25], 'o--',color='k',label="Layer=5")

plt.legend(loc="best")

plt.show()

plt.figure(2)
plt.title("Neural Network Training Time")
plt.xlabel("Number of Features")
plt.ylabel("Time(s)")

plt.grid()

x = [36, 76, 116, 156, 196]

plt.plot(x, z['mean_fit_time'][0:5], 'o--',color='r',label="Layer=1")
plt.plot(x, z['mean_fit_time'][5:10], 'o--',color='g',label="Layer=2")
plt.plot(x, z['mean_fit_time'][10:15], 'o--',color='b',label="Layer=3")
plt.plot(x, z['mean_fit_time'][15:20], 'o--',color='y',label="Layer=4")
plt.plot(x, z['mean_fit_time'][20:25], 'o--',color='k',label="Layer=5")
plt.legend(loc=1)

plt.show()

plt.figure(3)
plt.title("Neural Network Validation Accuracy")
plt.xlabel("Number of Features")
plt.ylabel("Accuracy")

plt.grid()

x = [36, 76, 116, 156, 196]
plt.plot(x, z['mean_test_score'][0:5], 'o-',color='r',label="Layer=1")
plt.plot(x, z['mean_test_score'][5:10], 'o-',color='g',label="Layer=2")
plt.plot(x, z['mean_test_score'][10:15], 'o-',color='b',label="Layer=3")
plt.plot(x, z['mean_test_score'][15:20], 'o-',color='y',label="Layer=4")
plt.plot(x, z['mean_test_score'][20:25], 'o-',color='k',label="Layer=5")
plt.legend(loc="best")

plt.show()

plt.figure(4)
plt.title("Neural Network Training Accuracy")
plt.xlabel("Number of Features")
plt.ylabel("Accuracy")

plt.grid()

x = [36, 76, 116, 156, 196]
plt.plot(x, z['mean_train_score'][0:5], 'o-',color='r',label="Layer=1")
plt.plot(x, z['mean_train_score'][5:10], 'o-',color='g',label="Layer=2")
plt.plot(x, z['mean_train_score'][10:15], 'o-',color='b',label="Layer=3")
plt.plot(x, z['mean_train_score'][15:20], 'o-',color='y',label="Layer=4")
plt.plot(x, z['mean_train_score'][20:25], 'o-',color='k',label="Layer=5")
plt.legend(loc="best")

plt.show()
```
